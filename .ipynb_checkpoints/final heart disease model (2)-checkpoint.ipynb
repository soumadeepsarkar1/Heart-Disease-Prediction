```python
# split data into feature & target data
x= df.iloc[:, :-1].values
y= df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(x,y, test_size=0.3, random_state=100)
```

```python
a3=[]
b3=[]
c3=[]
#using stochastic gradient descent 
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train= sc.fit_transform(X_train)
X_test= sc.transform(X_test)
```

```python
from sklearn.linear_model import SGDClassifier
n=np.linspace(100,5000)
a=np.array([])
for i in n:
    sgd=SGDClassifier(loss='log',max_iter=i, random_state=100)
    sgd.fit(X_train,y_train)
    a=np.append(a,sgd.score(X_test,y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(n,a)
plt.xlabel('no. of iterations')
plt.ylabel('Accuracy')
plt.show()
```

```python
sgd=SGDClassifier(loss='log',max_iter=5000, random_state=100)
sgd.fit(X_train,y_train)
sgd.score(X_train,y_train)
```

```python
pred=sgd.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
cm=confusion_matrix(y_test,pred)
print(confusion_matrix(y_test,pred))
```

```python
TN = cm[0][0]
TP = cm[1][1]
FN = cm[1][0]
FP = cm[0][1]
a=(TP+TN)/(TP+TN+FN+FP)
print("Accuracy: {}".format(a))
b=TP/(TP+FN)
print("Sensitivity: {}".format(b))
c=TN/(TN+FP)
print("Specificity: {}".format(c))
a3.append(round(a,3))
b3.append(round(b,3))
c3.append(round(c,3))
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
#Defining Hybrid Ensemble Learning Model
# create the sub-models
estimators = []
from sklearn.linear_model import LogisticRegression
#Defining 5 Logistic Regression Models

model11 = LogisticRegression()
estimators.append(('log1', model11))
model12 = LogisticRegression()
estimators.append(('log2', model12))
model13 = LogisticRegression()
estimators.append(('log3', model13))
model14 = LogisticRegression()
estimators.append(('log4', model14))
model15 = LogisticRegression()
estimators.append(('log5', model15))
```

```python
from sklearn.neighbors import KNeighborsClassifier
#Defining 5 K-NN classifiers
model16 = KNeighborsClassifier(n_neighbors = 1)
estimators.append(('knn1', model16))
model17 = KNeighborsClassifier(n_neighbors = 1)
estimators.append(('knn2', model17))
model18 = KNeighborsClassifier(n_neighbors = 1)
estimators.append(('knn3', model18))
model19 = KNeighborsClassifier(n_neighbors = 1)
estimators.append(('knn4', model19))
model20 = KNeighborsClassifier(n_neighbors = 1)
estimators.append(('knn5', model20))
```

```python
from sklearn.svm import SVC
#Defining 5 Support Vector Classifiers
model21 = SVC(kernel = 'rbf')
estimators.append(('svm1', model21))
model22 = SVC(kernel = 'rbf')
estimators.append(('svm2', model22))
model23 = SVC(kernel = 'rbf')
estimators.append(('svm3', model23))
model24 = SVC(kernel = 'rbf')
estimators.append(('svm4', model24))
model25 = SVC(kernel = 'rbf')
estimators.append(('svm5', model25))
```

```python
from sklearn.tree import DecisionTreeClassifier
#Defining 5 Decision Tree Classifiers
model26 = DecisionTreeClassifier()
estimators.append(('cart1', model26))
model27 = DecisionTreeClassifier()
estimators.append(('cart2', model27))
model28 = DecisionTreeClassifier()
estimators.append(('cart3', model28))
model29 = DecisionTreeClassifier()
estimators.append(('cart4', model29))
model30 = DecisionTreeClassifier()
estimators.append(('cart5', model30))
```

```python
from sklearn.naive_bayes import GaussianNB
#Defining 5 Naive Bayes classifiers
model31 = GaussianNB()
estimators.append(('nbs1', model31))
model32 = GaussianNB()
estimators.append(('nbs2', model32))
model33 = GaussianNB()
estimators.append(('nbs3', model33))
model34 = GaussianNB()
estimators.append(('nbs4', model34))
model35 = GaussianNB()
estimators.append(('nbs5', model35))
```

```python
# Defining the ensemble model
ensemble = VotingClassifier(estimators)
ensemble.fit(X_train, y_train)
y_pred = ensemble.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
#Confisuin matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
print(cm)
TN = cm[0][0]
TP = cm[1][1]
FN = cm[1][0]
FP = cm[0][1]
a=(TP+TN)/(TP+TN+FN+FP)
print("Accuracy: {}".format(a))
b=TP/(TP+FN)
print("Sensitivity: {}".format(b))
c=TN/(TN+FP)
print("Specificity: {}".format(c))
a3.append(round(a,3))
b3.append(round(b,3))
c3.append(round(c,3))
```

```python
print(classification_report(y_test,y_pred))
```

```python
#Cross-Validation
kfold = model_selection.KFold(n_splits=10, random_state=100,shuffle=True)
results = model_selection.cross_val_score(ensemble, X_train, y_train, cv=kfold)
print(results.mean())
```
